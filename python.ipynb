```python
import pandas as pd
import numpy as np
```

```python
arquivo_saida = open("arquivo_sem_cru.csv", "w")

with open("./taco-db-nutrientes.csv", "r")  as csv:
    linhas = csv.readlines()

linhas = [linha.split(",") for linha in linhas]

linhas_sem_cru = [linha for linha in linhas if "cru" not in linha[1]]

for linha in linhas_sem_cru:
    arquivo_saida.write(",".join(linha))
```
